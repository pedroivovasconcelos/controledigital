```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl

# Definir a funo de transferncia no plano Z
num = [0.5, 0]  # Numerador
den = [1, -0.5]  # Denominador

# Criar o sistema discreto
system = ctl.TransferFunction(num, den, 1)

# Plotar o lugar das razes
plt.figure()
ctl.root_locus(system, grid=True)
plt.title('Lugar das Razes no Plano Z')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginria')
plt.grid(True)

# Plotar o crculo unitrio
theta = np.linspace(0, 2*np.pi, 300)
x = np.cos(theta)
y = np.sin(theta)
plt.plot(x, y, 'k--', linewidth=1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl

# Definir a funo de transferncia no plano Z
num = [1, 1]  # Numerador
den = [1, 0.5]  # Denominador
system1 = ctl.TransferFunction(num, den, 1)
num = [1, 0.9]  # Numerador
den = [1, -0.5]  # Denominador
system2 = ctl.TransferFunction(num, den, 1)
num = [1]  # Numerador
den = [1, -0.9]  # Denominador
system3 = ctl.TransferFunction(num, den, 1)
num = [1]  # Numerador
den = [1, -1]  # Denominador
system4 = ctl.TransferFunction(num, den, 1)
system5 = system1*system2*system3*system4


# Plotar o lugar das razes
plt.figure()
ctl.root_locus(system5, grid=True)
plt.title('Lugar das Razes no Plano Z')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginria')
plt.grid(True)

# Plotar o crculo unitrio
theta = np.linspace(0, 2*np.pi, 300)
x = np.cos(theta)
y = np.sin(theta)
plt.plot(x, y, 'k--', linewidth=1)

plt.show()
```
